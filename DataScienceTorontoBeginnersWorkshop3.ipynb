```python
Cyclist = pd.read_csv("Cyclists.csv")
```

Here, I read the Cyclist data into a dataframe. This dataset is a subset of the Killed and Seriously Injured (KSI) dataset from 2007-2018 in Toronto. These events include any serious or fatal collision where a cyclist is involved.

```python
print (Cyclist.head())
```

I used the .head() function for dataframes to print out the first 5 rows of the Cyclist Dataset.

```python
x = Cyclist["YEAR"]
```

```python
plt.hist(x, bins=90)
plt.ylabel('No of People involved in Accidents')
plt.xlabel('Year')
plt.figure(figsize=(20,35))
plt.show()
```

This code is similar to the exercise I did with the Wards from before. Try to trace my code to figure out what I did!

```python
Year_Age = Cyclist.groupby(["YEAR", "INVAGE"])
```

Here, I use the groupby function to group the data frame by the year the accident occured in, as well as the incident victim's age.

```python
print(Year_Age.first())
```

By using the groupby function on a dataframe, I created a new data structure, called a "groupby object". Here, I outputted how the new object looks like.

```python
Y_A = Year_Age['X'].agg(['count'])
print (Y_A)
```

Here is the nice thing about groupby objects, we can apply functions to our data like the ".agg['count']" function I used which gave a count of the number of accident victims in each age group for each year we are looking at.

```python
fig, ax = plt.subplots(figsize=(15,7))
Y_A['count'].unstack().plot(ax=ax)
plt.ylabel("# of Accident Victims")
plt.show()
```

Finally, I created a line graph showing how the number of accidents varies per year with each line representing an age group.
