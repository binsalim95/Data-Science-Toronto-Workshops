```python
print(iris)
```

```python
data = iris.flatten()
```

The iris dataset contains data on the sepal length, sepal width, petal length & petal width on 50 flowers each belonging to the setosa, virginica & versicolor species. The goal of using this data set was to generate boxplots comparing the Sepal Length of each species. 

```python
Sepal_length = []
i=0
while i<len(data):
    if i%5==0:
        Sepal_length = np.append(Sepal_length,data[i])
    i=i+1

print (Sepal_length)
```

Here, I used a while loop to capture every 5th element in the data array since that corresponded to Sepal Length and brought it into a new, cleaner array. Try to Trace my code and see if you can figure out exactly how my while loop did it!

```python
Setosa = Sepal_length[0:50]
Versicolor = Sepal_length[50:100]
Virginica = Sepal_length[100:150]
```

Here, I used a feature common to Arrays, lists, tuples etc. Called "Slicing". Essentially this allows me to call all elements that fit within the bounds that I set. Ex. since the first 50 elements of the Sepal_length array corresponded to the Setosa, I sliced from 0 to 50 or [0:50] and so on for the other species.

```python
Plot = [Setosa, Versicolor, Virginica]
```

```python
fig = plt.figure(1, figsize=(9, 6))
ax = fig.add_subplot(111)
bp = ax.boxplot(Plot)
ax.set_xticklabels(['Setosa', 'Versicolor', 'Virginica'])
plt.show()
    
```
