Luckily for us, our data was ordered so we could use the formula for Median (https://en.wikipedia.org/wiki/Median). Given that we had an odd number of points, the formula for the median is the point corresponding to (n+1)/2 [where n = sample size]. 

For example, if say we have x = (2,5,9,20,25) we have n = 5 ordered points (Ordered means from least to greatest). The median would be (5+1)/2 point. In  this case, (5+1)/2 = 3 so the third point, 9, in x would be the median.

```python
z = np.random.rand(9)
print(z)
```

```python
def mean(x):
    n = len(x)
    A = 0
    for number in x:
        A = A + number
    return(A/n)

    
```

```python
def std(x):
    n = len(x)
    A = 0
    for number in x:
        A = A + (number-mean(x))**2
    return((A/n)**(1/2))
```

This is a function I wrote to find the Standard Deviation given any set of data. You can consult the powerpoint slides or google for the formula for Standard Deviation.

```python
std(x)
```

```python
data = pd.read_csv("Robbery_2014_to_2018.csv")
```

Here, I imported a data set on Robberies that Occurred in Toronto from 2014 - 2018. This data set was released by the Toronto Police. Found Here: http://data.torontopolice.on.ca/pages/open-data. One of the Goals of this workshop is to create a function that can model the number of robberies that Toronto recieves from year to year.

```python
Robbery = data.groupby("occurrenceyear")
Robbery.first()
```

```python
count = Robbery['X'].agg("count")
```

Here, I used the Groupby Function that I introduced during the Third Workshop to get a count of robberies per year, feel free to look over the notebook from that Workshop to understand the code above. Note that I named this Groupby object with the year and number of robberies as "count".

```python
print(count)
```

```python
x = []
y = []
for entry in count.iteritems():
    x.append(entry[0])
    y.append(entry[1])

print(x)
print(y)
del x[0:3]
del y[0:3]
print (x)
print (y)
```

Here, I used a For loop to split my data for the Years and the count of robberies into two different arrays. Using the .iteritems() function, I split my groupby object, count, into a list of lists that looked like this (Year, Count). So by looping over the entries in counts, I took the first number which corresponded to the year using entry[0] and added it to x and took the second entry using entry[1] and added it to y. 

If you noticed from before, there was actually an entry for robberies in 2004,2012 & 2013. Since the original data set is only for robberies from 2014-2018, I will treat those entries as outliers which I will get rid of using the del command. Above, I printed out what x and y looked like before and after I used the del command on them.

```python
plt.scatter(x,y)
plt.show()
```

Here is a plot of what the data looks like. As you can see, instead of it being linear, our data looks sinusoidal so let's try fitting a sinusoidal curve to it. The same steps below can be done using a linear model as scipy is able to help us optimize our models by minimizing the square error from our model to the data. 

```python
def fit_func(x,a,b,c,d):
    return (a*np.sin(c*(x-b))+d)
```

I defined the function, **a *sin(c*(x-b)) + d**, that we will be using to fit our data.

```python
X = np.linspace(1,5,5)
params = sci.curve_fit(fit_func,X,y)
a = params[0][0]
b = params[0][1]
c = params[0][2]
d = params[0][3]
print (a)
print (b)
print (c)
print (d)
```

Above, I used the scipy function, "curve_fit", to find parameters for the coefficients, a,b,c,d, such that square error between our estimate and the data would be minimized. 

```python
Ynew = []
for year in X:
    Ynew.append(int(fit_func(year, a, b,c,d)))
print (Ynew)
```

I ran the model over entries in X, to get the values the model would predict for the number of robberies in Toronto by Year and printed out what the model predicted as the number of robberies for each year.

```python
plt.plot(x,y, 'bo', x, Ynew, '--r')
plt.xlabel('Years')
plt.ylabel('Count')
plt.title ('Robberies in Toronto by Year')
plt.show()
```

Above is a plot of the actual count of robberies (in blue dots) and what our model predicted (with the red dashed line). As you can see, the model is still far from perfect, but does give a better fit than a linear model.

Going back to our data, we have many different features that we can choose from to do a multiple linear/polynomial regression with. For this notebook, I'll be making a model using the year and the day of the week to predict the number of robberies Toronto has.

```python
Robbery_Day = data.groupby(["occurrenceyear","occurrencedayofweek"])
Robbery_Day.first()
```

Above, I group the entries in our data by both their year and day of the week. Finally to obtain the number of robberies that occured for each available day and year combination, I'll use take a count.

```python
count = Robbery_Day['X'].agg("count")
print(count)
```

Now I have a count of the number of robberies for each Year+Day Combination.

```python
YearDay = []
Count = []
for entry in count.iteritems():
    if int(entry[0][0]) in (2014,2015,2016,2017,2018):
        YearDay.append(entry[0])
        Count.append(entry[1])
        

print(YearDay)

print(Count)
```

Remembering that we have outliers in our data since we have entries for years not between 2014 and 2018, I used a forloop to append the year/Day combination into 1 list and the corresponding count into another.

```python
YearDay = [list(entry) for entry in YearDay]
for entry in YearDay:
    if "Sunday" in entry[1]:
        entry[1] = 1
    elif "Monday" in entry[1]:
        entry[1] = 2
    elif "Tuesday" in entry[1]:
        entry[1] = 3
    elif "Wednesday" in entry[1]:
        entry[1] = 4
    elif "Thursday" in entry[1]:
        entry[1] = 5
    elif "Friday" in entry[1]:
        entry[1] = 6
    else:
        entry[1] = 7
        
Year = []
Day = []
for entry in YearDay:
    Year.append(entry[0])
    Day.append(entry[1])
print(Year)
print(Day)
```

Above, I relabelled my days. For Sunday I represented it with a 1 and so on untill I got to Saturday which is represented with a 7.

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_trisurf(Day, Year, Count, linewidth=0.2)
ax.view_init(30, 45)
ax.plot_trisurf(Day, Year, Count, cmap=plt.cm.jet, linewidth=0.01)
ax.set_xlabel('Days')
ax.set_ylabel('Years')
ax.set_zlabel('Count')
plt.rcParams["figure.figsize"] = (5,5)
plt.show()
```
